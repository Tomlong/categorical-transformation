```python
# Normalize data
def normalization(training_numpy, testing_numpy, columns):
    min_max = preprocessing.MinMaxScaler()
    min_max.fit(training_numpy)
    x_scaled = min_max.fit_transform(training_numpy)
    testting_x_scaled = min_max.transform(testing_numpy)
    training_norm = pd.DataFrame(x_scaled, columns = columns)
    testing_norm = pd.DataFrame(testting_x_scaled, columns = columns)
    return (training_norm.values, testing_norm.values)

def pca_function(n, data):
    pca = PCA(n_components=n)
    pca_data = pca.fit_transform(data)
    return (pca_data, pca)

def lda_function(X, y):
    lda = LDA()
    lda_data = lda.fit_transform(X, y)
    return (lda_data.flatten(), lda)

def replace_value(training_normal_numpy, key, loc):  
    training_normal_numpy[loc][key] = 1
```

```python
column_unuse = ['StartTime', 'SrcAddr', 'DstAddr', 'Sport']
column_str = ['Proto', 'State','Dir','Dport']
score_list = {}
fscore_list = {}

for index_train_dataset in range(3,4): 
    
    train = pd.read_csv('botnet_'+str(index_train_dataset)+'.binetflow')

    # Fill 0 to missing value
    train = train.fillna(0)    
    train = train[train['State'] != 0]

    # Drop unuse columns and row with missing value
    missing_state = []
    map(lambda word: missing_state.append(word) if word[-1]=='_' else None, np.unique(train['State']))
    
    for column in column_unuse:
        del train[column]
        
    for state in missing_state:
        train = train[train['State'] != state]
    

    # Transfer str label to int lable
    label_num = []
    
    map(lambda label: map(lambda key: label_num.append(key_str[key]) if key in label else None,key_str.keys()), train['Label'])
    train['Label'] = label_num
    

    # training & testing index
    train_index = np.where(train['Label'] == 0)[0].tolist() + np.where(train['Label'] == 1)[0].tolist()
    print 'training length:', len(train_index)

    # Numeric method
    state_numeric_dic = {}
    proto_numeric_dic = {}
    dir_numeric_dic = {}
    dport_numeric_dic = {}
    unique_state = Counter(train.iloc[train_index]['State'])
    unique_proto = Counter(train.iloc[train_index]['Proto'])
    unique_dir = Counter(train.iloc[train_index]['Dir'])
    unique_dport = Counter(train.iloc[train_index]['Dport'])

    for i,key in zip((np.random.choice(len(unique_state),len(unique_state),replace=False) + 1), unique_state.keys()):
        state_numeric_dic[key] = float(i)
    for i,key in zip((np.random.choice(len(unique_proto),len(unique_proto),replace=False) + 1), unique_proto.keys()):
        proto_numeric_dic[key] = float(i)
    for i,key in zip((np.random.choice(len(unique_dir),len(unique_dir),replace=False) + 1), unique_dir.keys()):
        dir_numeric_dic[key] = float(i)
    for i,key in zip((np.random.choice(len(unique_dport),len(unique_dport),replace=False) + 1), unique_dport.keys()):
        dport_numeric_dic[key] = float(i)

    state_numeric_array = []
    proto_numeric_array = []
    dir_numeric_array = []
    dport_numeric_array = []
    map(lambda value: state_numeric_array.append(state_numeric_dic[value]) if value in state_numeric_dic.keys() else state_numeric_array.append(0), train['State'])
    map(lambda value: proto_numeric_array.append(proto_numeric_dic[value]) if value in proto_numeric_dic.keys() else proto_numeric_array.append(0), train['Proto'])
    map(lambda value: dir_numeric_array.append(dir_numeric_dic[value]) if value in dir_numeric_dic.keys() else dir_numeric_array.append(0), train['Dir'])
    map(lambda value: dport_numeric_array.append(dport_numeric_dic[value]) if value in dport_numeric_dic.keys() else dport_numeric_array.append(0), train['Dport'])
    
    train_copy = pd.DataFrame()  
    for column in column_str:
        train_copy[column] = train[column]
        

    train_copy['state_numeric'] = state_numeric_array
    train_copy['proto_numeric'] = proto_numeric_array
    train_copy['dir_numeric'] = dir_numeric_array
    train_copy['dport_numeric'] = dport_numeric_array
   
    
    for dim_reduction in range(1,4):
    train_numeric_numpy_proto = normalization(train_copy.iloc[train_index]['proto_numeric'], ['Proto'])
    train_numeric_numpy_state = normalization(train_copy.iloc[train_index]['state_numeric'], ['State'])
    train_numeric_numpy_dir = normalization(train_copy.iloc[train_index]['dir_numeric'], ['Dir'])
    train_numeric_numpy_dport = normalization(train_copy.iloc[train_index]['dport_numeric'], ['Dport'])

    train_binary_proto = pd.get_dummies(train.iloc[train_index]['Proto'])
    train_binary_state = pd.get_dummies(train.iloc[train_index]['State'])
    train_binary_dir = pd.get_dummies(train.iloc[train_index]['Dir'])
    train_binary_dport= pd.get_dummies(train.iloc[train_index]['Dport'])
    
    train_binary_numpy_proto = normalization(train_binary_proto, train_binary_proto.columns)
    train_binary_numpy_state = normalization(train_binary_state, train_binary_state.columns)
    train_binary_numpy_dir = normalization(train_binary_dir, train_binary_dir.columns)
    train_binary_numpy_dport = normalization(train_binary_dport, train_binary_dport.columns)
    
    
    now_score = []
    now_fscore = []

    print 'Dataset ', index_train_dataset
    clf = knn(n_neighbors=1, algorithm='kd_tree')
    score = np.average(cross_val_score(clf, train_numeric_numpy_proto, np.array(label_num)[train_index], cv=5))
    fscore = np.average(cross_val_score(clf, train_numeric_numpy_proto, np.array(label_num)[train_index], cv=5, scoring='f1'))
    print 'numeric_proto: ',score, fscore    
    now_score.append(score)
    now_fscore.append(fscore)

    score = np.average(cross_val_score(clf, train_numeric_numpy_state, np.array(label_num)[train_index], cv=5))
    fscore = np.average(cross_val_score(clf, train_numeric_numpy_state, np.array(label_num)[train_index], cv=5, scoring='f1'))
    print 'numeric_state: ',score, fscore    
    now_score.append(score)
    now_fscore.append(fscore)
    
    score = np.average(cross_val_score(clf, train_numeric_numpy_dir, np.array(label_num)[train_index], cv=5))
    fscore = np.average(cross_val_score(clf, train_numeric_numpy_dir, np.array(label_num)[train_index], cv=5, scoring='f1'))
    print 'numeric_dir: ',score, fscore    
    now_score.append(score)
    now_fscore.append(fscore)
    
    score = np.average(cross_val_score(clf, train_numeric_numpy_dport, np.array(label_num)[train_index], cv=5))
    fscore = np.average(cross_val_score(clf, train_numeric_numpy_dport, np.array(label_num)[train_index], cv=5, scoring='f1'))
    print 'numeric_dport: ',score, fscore    
    now_score.append(score)
    now_fscore.append(fscore)
    
    score = np.average(cross_val_score(clf, train_binary_numpy_proto, np.array(label_num)[train_index], cv=5))
    fscore = np.average(cross_val_score(clf, train_binary_numpy_proto, np.array(label_num)[train_index], cv=5, scoring='f1'))
    print 'binary_proto: ',score, fscore    
    now_score.append(score)
    now_fscore.append(fscore)
    
    score = np.average(cross_val_score(clf, train_binary_numpy_state, np.array(label_num)[train_index], cv=5))
    fscore = np.average(cross_val_score(clf, train_binary_numpy_state, np.array(label_num)[train_index], cv=5, scoring='f1'))
    print 'binary_state: ',score, fscore    
    now_score.append(score)
    now_fscore.append(fscore)
    
    score = np.average(cross_val_score(clf, train_binary_numpy_dir, np.array(label_num)[train_index], cv=5))
    fscore = np.average(cross_val_score(clf, train_binary_numpy_dir, np.array(label_num)[train_index], cv=5, scoring='f1'))
    print 'binary_dir: ',score, fscore    
    now_score.append(score)
    now_fscore.append(fscore)
    
    score = np.average(cross_val_score(clf, train_binary_numpy_dport, np.array(label_num)[train_index], cv=5))
    fscore = np.average(cross_val_score(clf, train_binary_numpy_dport, np.array(label_num)[train_index], cv=5, scoring='f1'))
    print 'binary_dport: ',score, fscore    
    now_score.append(score)
    now_fscore.append(fscore)
    
    score_list[str(index_train_dataset)] = now_score
    fscore_list[str(index_train_dataset)] = now_fscore
    #print 'finish'
```
